```python
#Zadanie 1 
#Wgraj zawartość pliku orders.csv do ramki danych. 
#Przyjrzyj się danym, wykorzystując poznane metody (describe, info, head) zanalizuj dane.
import pandas as pd
import numpy as np
import csv 
dane = pd.read_csv('/content/drive/MyDrive/PAD Zadania 4/orders.csv', sep =';')
df = pd.DataFrame(dane)
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.head()
```

```python
#Zamień order_date na format daty. 
#Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;
df['order_date'] = pd.to_datetime(df['order_date'])
df
```

```python
# Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;
df['tshirt_category'].value_counts()
```

```python
#Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje 
#jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;
df['tshirt_category'] = df.tshirt_category.apply(lambda x: str(x).replace('Wh ', 'White '))
df['tshirt_category'] = df.tshirt_category.apply(lambda x: str(x).replace('Bl ', 'Black '))
df['tshirt_category'] = df.tshirt_category.apply(lambda x: str(x).replace('Tshirt ', 'T-Shirt '))
df
```

```python
df['tshirt_category'].value_counts()
```

```python
#Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, !dodaj apply
#która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.
df['tshirt_category'] = df.tshirt_category.apply(lambda x: str(x).replace('Tennis Shirt', 'TennisShirt'))
```

```python
df[['tshirt_colour','tshirt_type','tshirt_size']] = df['tshirt_category'].str.split(expand = True)
df
```

```python
df['tshirt_colour'].value_counts()
```

```python
def groups(x):
    if x == "Hoodie":
        return 'Hoodie'
    elif x == 'TennisShirt':
        return 'Tennis Shirt'
    else:
      return 'T-Shirt'
df['tshirt_type'] = df['tshirt_colour'].apply(groups)

```

```python
def groups_2(x):
    if x == "Hoodie":
        return np.NaN
    elif x == 'TennisShirt':
        return np.NaN
    elif x =='T-Shirt':
      return np.NaN
    else:
      return x
df['tshirt_colour'] = df['tshirt_colour'].apply(groups_2)
```

```python
df['tshirt_category'] = df.tshirt_category.apply(lambda x: str(x).replace('TennisShirt', ' Tennis Shirt'))
```

```python
df['tshirt_category'].unique()
```

```python
#Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016
mask = (df['order_date'] >= "2014-12-31") & (df['order_date'] < "2016-08-02")
df.order_id[mask]
```
